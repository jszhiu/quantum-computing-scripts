```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Create monitor input for I and Q values (siginput1 and siginput2)

length = 4096
#Iones = np.ones(4016)
Iones = np.full(4075, 736e-3)
Izeros = np.zeros(length-4075)
I = np.concatenate((Iones, Izeros))
Q = np.concatenate((Iones, Izeros))
sample = np.arange(0, length, 1)

plt.plot(sample, I)
```

```python
## Create weight function
smpFreq = 1.8e9
freq = 50e3
period = 1/freq

samplesCos = period*smpFreq

freqaxis = np.linspace(-(2*np.pi), 2*np.pi, int(samplesCos*1.75)) #multiplied by 1.75 as np.cos function creates 1.75 of a cosine wave

weight1 = np.cos(freqaxis)

plt.plot(freqaxis[:4096], weight1[:4096])
```

```python
## Code weighted integration formula
Vi = np.zeros(4096)
Vi[0] = weight1[0] * I[0]
for n in range(1, 4096):
    Vi[n] = weight1[n] * I[n]

print('The integration value is : %.2f V' % np.sum(Vi))
```
